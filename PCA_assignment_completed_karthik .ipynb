```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
df = pd.read_csv("wine.csv")
df.describe()
df.head()
```

```python
df.data = df.iloc[:,1:]
df.data.head()
# Converting into numpy array
DF = df.data.values
DF
```

```python
# Normalizing the numerical data 
df_normal = scale(DF)
```

```python
df_normal
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(df_normal)
pca_values 
```

```python
#loadings or weights
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:8],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8']), df[['Type']]], axis = 1)
finalDf
```
